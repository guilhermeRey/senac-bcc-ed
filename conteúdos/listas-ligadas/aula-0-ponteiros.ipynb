## Aula 1 - Revisão de Ponteiros

Vamos relembrar os ponteiros em C 😬. Lembram da definição de ponteiros?

## Conceito de ponteiros

**Ponteiros são variáveis que armazenam endereços de memória de outras variáveis**

## 1. Como enxergar o endereço de memória

Um exemplo simples para lembrarmos:

````c
int x = 10;
printf("Valor de x: %d\n", x);
printf("Endereço de x: %p\n", (void*)&x);
````

O especificador `%p` é utilizado para imprimir endereços de memória em formato hexadecimal. O cast para `(void*)` é uma boa prática para garantir portabilidade, uma vez que `%p` espera um ponteiro genérico.

O endereço impresso será algo como `0x7ffeeb3a4c5c` — um valor hexadecimal que representa a posição na memória. Esse valor pode variar entre execuções, pois o sistema operacional é responsável pela alocação de memória.

Entendemos então que um **ponteiro** é uma variável cujo valor é o **endereço de memória** de outra variável.

A declaração:

```c
int *p;
```

Define `p` como um ponteiro para `int` — ou seja, `p` é capaz de armazenar o endereço de memória de uma variável do tipo `int`.

## 2. Operadores fundamentais


| Operador | Função | Descrição |
|----------|--------|-----------|
| `&x` | Endereço-de | Retorna o endereço de memória da variável `x` |
| `*p` | Desreferenciação (indireção) | Acessa o valor armazenado no endereço contido em `p` |

Esses dois operadores são complementares:

- `&` parte de uma variável e obtém seu endereço
- `*` parte de um endereço e obtém o valor armazenado nele

````c
int x = 10;
int *p = &x;  //p armazena o endereço de x

printf("Valor de x: %d\n", x);              //10
printf("Endereço de x: %p\n\n", (void*)&x);   //endereço de x
printf("Valor de p: %p\n", (void*)p);       //mesmo endereço
printf("Valor apontado por p: %d\n", *p);   //10
printf("Edereço de p: %p\n", (void*)&p);   //10
````

Neste exemplo, `p` e `&x` contêm o mesmo valor: o endereço de `x`. A expressão `*p` acessa o conteúdo armazenado nesse endereço, retornando o valor de `x`.

Uma consequência importante é que **é possível modificar o valor de `x` por meio do ponteiro**:

````c
int x = 10;
int *p = &x;  //p armazena o endereço de x
*p = 99;
printf("Valor de x: %d\n", x);  //99
````

## 3. O valor `NULL`

A constante `NULL` representa um ponteiro que **não aponta para nenhum endereço válido**. É definida no cabeçalho `<stddef.h>` (e incluída indiretamente por `<stdio.h>` e `<stdlib.h>`).

````c
int *p = NULL;
if (p == NULL) {
    printf("Ponteiro nulo\n");
} 
else {
    printf("Valor: %d\n", *p);
}
````

## 4. Funções: passagem por valor e passagem por referência

Em C, todos os argumentos são passados **por valor**. Isso significa que a função recebe uma cópia do argumento, e qualquer modificação feita sobre essa cópia não afeta a variável original:

````c
#include <stdio.h>

void incrementa(int x) {
    x++;  //modifica apenas a cópia local
}

int main() {
    int n = 5;
    incrementa(n);
    printf("%d\n", n);  //imprime 5 (inalterado)
}
````

Mas existe uma maneira de alterarmos o valor através da função diretamente, como vimos quando entendemos que é possível mexer em um valor de uma variável pelo ponteiro: **passamos o parâmetro por referência** (que, na verdade, é só um endereço de memória sendo passado por valor).

````c
#include <stdio.h>

void incrementa(int *p) {
    (*p)++;  //modifica o valor no endereço original
}

int main() {
    int n = 5;
    incrementa(&n);
    printf("%d\n", n);  //imprime 6
}
````

## 5. Arrays e Ponteiros

Em C, existe uma relação estreita entre arrays e ponteiros. Em muitos contextos, o nome de um array "decai" (*decays*) para um ponteiro para o seu primeiro elemento:

````c
#include<stdio.h>

int v[3] = {10, 20, 30};
int *p = v;  //equivalente a &v[0]
printf("v[0] = %d\n", *p);
printf("endereço de v[0] = %p", (void*)p);
````

### Aritmética de ponteiros

Quando um ponteiro é incrementado, ele avança o número de bytes correspondente ao tamanho do tipo apontado:

- `p + 0` aponta para `v[0]`
- `p + 1` aponta para `v[1]`
- `p + 2` aponta para `v[2]`

Consequentemente:

- `*(p + i)` é equivalente a `v[i]`

Na realidade, a notação de indexação `v[i]` é definida pelo padrão C como equivalente a `*(v + i)`. Trata-se de açúcar sintático (*syntactic sugar*).

Você pode percorrer um vetor usando ponteiros dessa maneira:

````c
#include<stdio.h>

int v[3] = {10, 20, 30};
for (int i = 0; i < 3; i++) {
    printf("%d ", *(v + i));
}
````

## 6. Strings

Em C, não existe um tipo específico para strings. Uma string é representada como um **array de `char`** terminado pelo caractere nulo `'\0'`.

```c
char s[] = "casa";
```

A representação em memória é:

```
| 'c' | 'a' | 's' | 'a' | '\0' |
```

O caractere `'\0'` sinaliza o fim da string. Sem ele, funções como `printf`, `strlen` e `strcpy` continuariam lendo posições de memória além dos limites da string.

Uma string pode ser percorrida com ponteiro:

````c
char s[] = "casa";
char *p = s;
while (*p != '\0') {
    putchar(*p);
    p++;
}
````

## 7. Operador `->`

Ao trabalhar com ponteiros para estruturas, utiliza-se o operador `->` para acessar os campos da estrutura apontada.

Considere a seguinte definição:

```c
typedef struct {
    char nome[50];
    int idade;
} Pessoa;
```

O acesso aos campos pode ser realizado de duas formas:

````c
typedef struct {
    char nome[50];
    int idade;
} Pessoa;

Pessoa maria;
maria.idade = 25;            //acesso direto (variável)

Pessoa *p = &maria;
p->idade = 25;               //acesso via ponteiro
//equivalente a: (*p).idade = 25
````

O operador `->` é equivalente à desreferenciação seguida do acesso com ponto (`(*p).campo`), porém com sintaxe mais concisa e legível. Na prática, é a forma padrão de acessar campos de uma estrutura por meio de ponteiro.

## 8. Alocação de memória

As variáveis locais residem na **pilha** (*stack*) e são automaticamente destruídas ao fim do escopo em que foram declaradas. Para criar dados que persistam além do escopo, ou cujo tamanho só é determinado em tempo de execução, utiliza-se a **alocação dinâmica** (*heap*).

````c
int *p = malloc(sizeof(int));
if (p == NULL) {
    printf("Erro de alocação\n");
    return 1;
}
*p = 123;
printf("Valor: %d\n", *p);
free(p);
````

### Regras fundamentais

1. **Toda memória alocada com `malloc` deve ser liberada com `free`.** A omissão dessa liberação resulta em **vazamento de memória** (*memory leak*).

2. **Após `free(p)`, o ponteiro `p` não deve ser utilizado.** O endereço armazenado em `p` deixa de ser válido, e qualquer acesso constitui comportamento indefinido.

3. **Recomendação:** após `free(p)`, atribuir `p = NULL;`. Essa prática facilita a detecção de erros, pois qualquer tentativa subsequente de desreferenciação resultará em um erro previsível.
