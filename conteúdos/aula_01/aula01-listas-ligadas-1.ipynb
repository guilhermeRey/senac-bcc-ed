### Célula (ou nó)

Vamos começar representando um nó em C, através de `structs`.

````c
#include<stdio.h>

struct no {
    int info;         // informação do nó
    struct no * prox; // ponteiro para o próximo nó. É auto-referência!
};
````

Cada nó da lista tem:
- `info`: o valor armazenado
- `prox`: um ponteiro para o próximo nó da lista

Dessa maneira podemos começar a brincar com listas ligadas.

Como podemos criar uma lista que seja `1 -> 2 -> 3`, por exemplo?

Precisamos criar um ponteiro para ser o início da lista.

```c
struct no * lista;
```

### É possível criarmos uma lista de maneira simples, como variáveis?

````c
#include <stdio.h>

struct no {
    int info;
    struct no *prox;
};

int main() {
    struct no n1, n2, n3;

    n1.info = 1;
    n1.prox = &n2;

    n2.info = 2;
    n2.prox = &n3;

    n3.info = 3;
    n3.prox = NULL;

    struct no *lista = &n1;
    printf("%d", lista->info);
}
````

Isso funciona, mas não é tão interessante. Listas devem ser dinâmicas, nessa abordagem nós temos que definir a quantidade de nós para iniciarmos.

### Então como podemos fazer de maneira dinâmica?

````c
#include<stdio.h>

struct no {
    int info;         
    struct no * prox;
};

int main() {
    struct no * lista = malloc(sizeof(struct no));
    lista->info = 1;
    lista->prox = malloc(sizeof(struct no));

    lista->prox->info = 2;
    lista->prox->prox = malloc(sizeof(struct no));

    lista->prox->prox->info = 3;
    lista->prox->prox->prox = NULL;

    printf("%d\n", lista->info);
    printf("%d\n", lista->prox->info);
    printf("%d\n", lista->prox->prox->info);
    // printf("%d\n", lista->prox->prox->prox->info);
}
````

### Antes de continuar, como imprimir os valores?

Por enquanto nós estamos imprimindo os valores dessa maneira:

```c
printf("%d\n", lista->info);
printf("%d\n", lista->prox->info);
printf("%d\n", lista->prox->prox->info);
```

Nem sempre vamos saber a quantidade de nós em uma lista. Como percorrer a lista inteira e imprimir seus valores?

````c
#include<stdio.h>

struct no {
    int info;         
    struct no * prox;
};

int main() {
    struct no * lista = malloc(sizeof(struct no));
    lista->info = 1;
    lista->prox = malloc(sizeof(struct no));

    lista->prox->info = 2;
    lista->prox->prox = malloc(sizeof(struct no));

    lista->prox->prox->info = 3;
    lista->prox->prox->prox = NULL;

    struct no * curr = lista; // criando uma variável para termos o inicio da lista
    while (curr != NULL) {
        printf("%d ", curr->info);
        curr = curr->prox;
    }
}
````

### Legal! Mas ainda assim isso parece esquisito...

A criação dessa maneira não é muito útil. Imagine se tivermos 100 nós, faremos 100 `->prox` no código!

Por isso, precisamos criar funções auxiliares. Começamos com uma simples: a criação de um ponteiro de nó.

````c
#include<stdio.h>

struct no {
    int info;         
    struct no * prox;
};

struct no * novoNo(int info) {
    struct no * novo = malloc(sizeof(struct no));
    novo->info = info;
    return novo;
}

int main() {
    struct no * lista = novoNo(1);
    lista->prox = novoNo(2);
    lista->prox->prox = novoNo(3);
    // Finalizando a lista
    lista->prox->prox->prox = NULL;

    struct no * curr = lista; // criando uma variável para termos o inicio da lista
    while (curr != NULL) {
        printf("%d ", curr->info);
        curr = curr->prox;
    }
}
````

Já melhorou! Agora não temos `malloc`s por ai. Mas ainda assim, não resolvemos nosso problema principal: montes de `->prox->prox` para criar novos nós na lista.

### Será que podemos melhorar? Como inserir um elemento na lista?

Precisamos fazer com que inserir novos elementos seja simples, algo como `inserir(10)` ou algo parecido. 

Como estamos usando a linguagem C, precisamos passar o parâmetro que determinar onde inicia a lista e o valor que queremos inserir.

A maneira mais simples de fazermos isso é inserindo no início da lista! Vamos implementar `insert_first(list, value)`

````c
#include<stdio.h>

struct no {
    int info;         
    struct no * prox;
};

struct no * novoNo(int info) {
    struct no * novo = malloc(sizeof(struct no));
    novo->info = info;
    return novo;
}

struct no * insert_first(struct no * lista, int info) {
    struct no *novo = novoNo(info);
    if (!novo) return lista;   // se falhar, mantém a lista como estava
    novo->prox = lista;        // novo aponta para a antiga cabeça
    return novo;               // novo vira a cabeça
}

int main() {
    struct no * lista = novoNo(1);
    lista = insert_first(lista, 2);
    lista = insert_first(lista, 3);

    struct no * curr = lista; // criando uma variável para termos o inicio da lista
    while (curr != NULL) {
        if (curr->prox != NULL)
            printf("%d...aponta para (%d)\n", curr->info, curr->prox->info);
        else
            printf("%d\n", curr->info);
        
        curr = curr->prox;
    }
}
````

Por que `insert_first` é do tipo `struct no *`?

### Como podemos mudar a ordem de inserção?

Será que podemos fazer uma inserção no final? O que isso implica?

````c
#include<stdio.h>

struct no {
    int info;         
    struct no * prox;
};

struct no * novoNo(int info) {
    struct no * novo = malloc(sizeof(struct no));
    novo->info = info;
    return novo;
}

struct no * insert_first(struct no * lista, int info) {
    struct no *novo = novoNo(info);
    if (!novo) return lista;   // se falhar, mantém a lista como estava
    novo->prox = lista;        // novo aponta para a antiga cabeça
    return novo;               // novo vira a cabeça
}

void insert_last(struct no * lista, int info) {
    struct no *novo = novoNo(info);
    // Precisamos encontrar o último elemento da lista. Quando encontramos, adicionamos o novo!
    struct no *curr = lista;
    while (curr->prox != NULL) {
        curr = curr->prox;
    }
    curr->prox = novo;
}

int main() {
    struct no * lista = novoNo(1);
    insert_last(lista, 2);
    insert_last(lista, 3);

    struct no * curr = lista; // criando uma variável para termos o inicio da lista
    while (curr != NULL) {
        if (curr->prox != NULL)
            printf("%d...aponta para (%d)\n", curr->info, curr->prox->info);
        else
            printf("%d\n", curr->info);
        
        curr = curr->prox;
    }
}
````

Por que `insert_first` é do tipo `struct no *` e `insert_last` é do tipo `void`?

### Aprendemos a inserir. E como remover?

````c
#include<stdio.h>

struct no {
    int info;         
    struct no * prox;
};

struct no * novoNo(int info) {
    struct no * novo = malloc(sizeof(struct no));
    novo->info = info;
    return novo;
}

struct no * insert_first(struct no * lista, int info) {
    struct no *novo = novoNo(info);
    if (!novo) return lista;   // se falhar, mantém a lista como estava
    novo->prox = lista;        // novo aponta para a antiga cabeça
    return novo;               // novo vira a cabeça
}

void insert_last(struct no * lista, int info) {
    struct no *novo = novoNo(info);
    // Precisamos encontrar o último elemento da lista. Quando encontramos, adicionamos o novo!
    struct no *curr = lista;
    while (curr->prox != NULL) {
        curr = curr->prox;
    }
    curr->prox = novo;
}

struct no * remove_first(struct no * lista) {
    if (lista == NULL) return NULL; // lista vazia

    struct no *novo_inicio = lista->prox;
    return novo_inicio;
}

int main() {
    struct no * lista = novoNo(1);
    insert_last(lista, 2);
    insert_last(lista, 3);

    struct no * curr = lista; // criando uma variável para termos o inicio da lista
    while (curr != NULL) {
        if (curr->prox != NULL)
            printf("%d...aponta para (%d)\n", curr->info, curr->prox->info);
        else
            printf("%d\n", curr->info);
        
        curr = curr->prox;
    }

    printf("\n\n");
    
    lista = remove_first(lista);
    curr = lista; // criando uma variável para termos o inicio da lista
    while (curr != NULL) {
        if (curr->prox != NULL)
            printf("%d...aponta para (%d)\n", curr->info, curr->prox->info);
        else
            printf("%d\n", curr->info);
        
        curr = curr->prox;
    }
}
````
